```python
dataset['Sex'] = dataset['Sex'].replace(to_replace = {'male': 1, 'female': 0})
dataset['Embarked'] = dataset['Embarked'].replace(to_replace = {'S': 1, 'C': 2, 'Q': 3})
```

```python
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].mean())
```

```python
def normalization(data):
    data_range = data.max() - data.min()
    avg = data.mean()
    data = (data - avg) / data_range
    return data
```
